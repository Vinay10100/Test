```python
import numpy as np
import random
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
import os,cv2
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import layers, optimizers
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
def read_and_preprocess(img_path):
    try:
        img = cv2.imread(img_path, cv2.IMREAD_COLOR) # reading the image
        img = cv2.resize(img, (256, 256)) # resizing it 
        img = np.array(img, dtype='float32') # convert its datatype so that it could be normalized
        img = img/255 # normalization (now every pixel is in the range of 0 and 1)
        return img
    except:
        print(f"Error reading image: {img_path}")
        return None
    
```

```python
X_train = []
y_train = []
train_path = 'C:/Users/VINAY/Desktop/yamaha hackathon/Dataset/Train/'

for folder in os.scandir(train_path):
    if folder.is_dir():
        class_label = folder.name
        for entry in os.scandir(folder.path):
            if entry.is_file() and entry.name.endswith('.jpg'):
                img_path = os.path.join(folder.path, entry.name)
                print('Reading image:', img_path)
                img = read_and_preprocess(img_path)
                if img is not None:
                    X_train.append(img)
                    if class_label == 'Type A':
                        y_train.append(0)
                    elif class_label == 'Type B':
                        y_train.append(1)
                    else:
                        y_train.append(2)

X_train = np.array(X_train)
y_train = np.array(y_train)
```

```python
sns.set(style='darkgrid')
sns.countplot(x=y_train)
plt.xticks(ticks=[0, 1, 2], labels=['Type A', 'Type B', 'Type C'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Define the number of images to display per class
n_images = 3

# Define a dictionary to map class labels to string names
class_names = {0: 'Type A', 1: 'Type B', 2: 'Type C'}

# Create a figure with subplots
fig, axes = plt.subplots(nrows=3, ncols=n_images, figsize=(10, 6))

# Loop through each class and select n_images random examples
for i in range(3):
    class_indices = np.where(np.array(y_train) == i)[0]
    random_indices = np.random.choice(class_indices, n_images, replace=False)
    for j, idx in enumerate(random_indices):
        # Load and display the image
        img = X_train[idx]
        axes[i, j].imshow(img)
        axes[i, j].axis('off')
        # Set the title to the class name
        axes[i, j].set_title(class_names[i])

plt.suptitle('Sample Images per Class', fontsize=16)
plt.show()
```

```python
#printing Horizontaly Flipped Image
X_new = np.fliplr(X_train[0])
plt.imshow(X_new)
```

```python
X_aug = []
y_aug = []

for i in range(0, len(y_train)):
    X_new = np.fliplr(X_train[i])
    X_aug.append(X_new)
    y_aug.append(y_train[i])
```

```python
X_aug = np.array(X_aug)
y_aug = np.array(y_aug)
```

```python
X_train = np.append(X_train, X_aug, axis=0) # appending augmented images to original training samples
X_train.shape
```

```python
y_train = np.append(y_train, y_aug, axis=0)
y_train.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15, shuffle=True, stratify=y_train, random_state=123)
```

```python
y_train.shape
```

```python
y_val.shape
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential([
    #input layer
    Conv2D(filters=32, kernel_size=(2,2), activation='relu', input_shape=(256, 256, 3)),
    MaxPooling2D((4,4)),
    
    Conv2D(filters=64, kernel_size=(3,3), activation='relu', padding='same'),
    MaxPooling2D((3,3)),
    Dropout(0.3), # for regularization to prevent overfitting
    
    Conv2D(filters=64, kernel_size=(4,4), activation='relu', padding='same'),
    Conv2D(filters=128, kernel_size=(5,5), activation='relu', padding='same'),
    MaxPooling2D((2,2)),
    Dropout(0.4),
    
    Conv2D(filters=128, kernel_size=(5,5), activation='relu', padding='same'),
    MaxPooling2D((2,2)),
    Dropout(0.5),
    
    Flatten(), # flattening reshapes the output from the convolutional layers into a 1D array
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dense(3, activation='softmax')
])
# Slowing down the learning rate
opt = optimizers.Adam(learning_rate=0.0001)
# compile the model
model.compile(loss = 'sparse_categorical_crossentropy', optimizer=opt, metrics= ["accuracy"])
# use early stopping to exit training if validation loss is not decreasing even after certain epochs (patience)
earlystopping = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)

# save the best model with least validation loss
checkpointer = ModelCheckpoint(filepath="Classifier.h5", verbose=1, save_best_only=True)
history = model.fit(X_train, y_train, epochs = 100, validation_data=(X_val, y_val), batch_size=32, shuffle=True, callbacks=[earlystopping, checkpointer])
```

```python
test_acc = model.evaluate(X_train, y_train)
```

```python
print('accuracy:', test_acc)
```

```python
#plot training loss and validation loss 
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```python
#plot training accuracy and validation accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Load pretrained model
with open('Classifier_main.json', 'r') as json_file:
    json_savedModel= json_file.read()
# load the model  
model = tf.keras.models.model_from_json(json_savedModel)
model.load_weights('Classifier.h5')
model.compile(loss = 'sparse_categorical_crossentropy', optimizer=opt, metrics= ["accuracy"])
```

```python
X_test = []
y_test = []
test_path = 'C:/Users/VINAY/Desktop/yamaha hackathon/Dataset/Test/'

for folder in os.scandir(test_path):
    if folder.is_dir():
        class_label = folder.name
        for entry in os.scandir(folder.path):
            if entry.is_file() and entry.name.endswith('.jpg'):
                img_path = os.path.join(folder.path, entry.name)
                print('Reading image:', img_path)
                img = read_and_preprocess(img_path)
                if img is not None:
                    X_test.append(img)
                    if class_label == 'Type A':
                        y_test.append(0)
                    elif class_label == 'Type B':
                        y_test.append(1)
                    else:
                        y_test.append(2)

X_test = np.array(X_test)
y_test = np.array(y_test)
```

```python
sns.set(style='darkgrid')
sns.countplot(x=y_test)
plt.xticks(ticks=[0, 1, 2], labels=['Type A', 'Type B', 'Type C'])
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# making predictions
predictions = model.predict(X_test)
```

```python
predictions.shape
```

```python
# Obtain the predicted class from the model prediction
predict = [] #list

for i in predictions:
  predict.append(np.argmax(i))

predict = np.asarray(predict)
```

```python
# Obtain the accuracy of the model
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predict)
accuracy
```

```python
# plot the confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predict)
plt.figure(figsize = (7,7))
sns.heatmap(cm, annot=True, cmap='Blues')

# The model misclassified one Covid case as Normal
```
